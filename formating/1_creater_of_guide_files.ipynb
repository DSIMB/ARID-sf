```python
import sys, os
import pandas as pd
import numpy as np
```

Creates a guide file for formatting antibody antigen complexes to ARID-sf usable format.

The guide file is a tabulated separated value file with the the path of the .pdb model you want to convert, the antigen chains and the antibody chains.

See code example below

To format ONE system:
The scripts are to be executed in order (1_ to 4_)

1_:\
    produces the guidefile.csv

2_:\
    creates the haddock parameter file and required structures in path_to/ensemble.pdb (concatenate all structures into one file)
    Here anything can be added to the .cfg file, for example if you wish to minimize the structure [emscoring] can be added...
    python3 2_model_to_ua.py path_to/ensemble.pdb guidefile.csv number_of_cpus

3_:\
    launch HADDOCK3 for topology conversion
    python3 3_launch_haddock_runs.py path_to_CFG_directory

4_:\
    parse haddock results and copy files
    python parse_haddock_top.py path_haddock_topol path_models_output
    don't forget to clean the haddock directories!


```python
# references from ADB
adbrefs_path = 'path_to/input_structures/'
adbrefs = [os.path.join(adbrefs_path,f) for f in os.listdir(adbrefs_path) if not (f.endswith('_AG.pdb') or f.endswith('_AB.pdb'))]

guide_construct = []
for f in adbrefs:
    guide_construct.append({
        'PDBpath':f,
        'chains_AG':'A',
        'chains_AB':'B'
    })
guide_construct = pd.DataFrame(guide_construct)
guide_construct.to_csv('path_to/guideADB_refs.csv',index=False,sep='\t')
```

```python
# rosetta produced files
path_files = 'path_to/Covid2019'
# Abs : L,H
# Ags : C
all_productions = [f for f in os.listdir(path_files) if f.startswith('PRODUCTION')]
for prod in all_productions:
    path_prod = os.path.join(path_files,prod)
    guide_construct = []
    for pdb in os.listdir(path_prod):
        if not pdb.endswith('.pdb'):
            continue
        path_pdb = os.path.join(path_prod,pdb)
        guide_construct.append({
        'PDBpath':path_pdb,
        'chains_AG':'C',
        'chains_AB':'L,H' # <== if multiple chains, separate with ","
    })
    guide_construct = pd.DataFrame(guide_construct)
    guide_construct.to_csv('/pathto/guideRosetta{}.csv'.format(prod),index=False,sep='\t')
```
